```python
import numpy as np
import matplotlib.pyplot as plt
from qiskit import QuantumCircuit
from qiskit_aer import Aer
from qiskit.visualization import array_to_latex
```

```python
circ = QuantumCircuit(1)
circ.draw('mpl')
```

```python
circ.x(0)
circ.y(0)
circ.draw('mpl')
```

```python
state_sim = Aer.get_backend('statevector_simulator')
result = state_sim.run(circ).result()
statevec = result.get_statevector()

array_to_latex(statevec)
```

```python
circ = QuantumCircuit(1)
circ.x(0)
circ.h(0)
circ.measure_all()
circ.draw('mpl')
```

```python
qasm_sim = Aer.get_backend('qasm_simulator') # Aqui estamos simulando as medi√ß√µes!
result = qasm_sim.run(circ, shots = 1000000).result()
counts = result.get_counts()

from qiskit.visualization import plot_histogram
plot_histogram(counts)
```

```python
circ = QuantumCircuit(2)
reg = ClassicalRegister(2)
circ.add_register(reg)

circ.x(0)
# Aqui, o 1¬∫ √© o qubit de controle e o 2¬∫ o alvo
circ.cx(0, 1)

circ.measure(0,0)
circ.measure(1,1)

circ.draw('mpl')
```

```python
def my_run(circ):
    qasm_sim = Aer.get_backend('qasm_simulator') # Aqui estamos simulando as medi√ß√µes!
    result = qasm_sim.run(circ, shots = 10000).result()
    counts = result.get_counts()
    return counts

counts = my_run(circ)

plot_histogram(counts)
```

```python
circ = QuantumCircuit(2)
reg = ClassicalRegister(2)
circ.add_register(reg)

circ.h(0)
# Aqui, o 1¬∫ √© o qubit de controle e o 2¬∫ o alvo
circ.cx(0, 1)

circ.measure(0,0)
circ.measure(1,1)

circ.draw('mpl')
```

```python
counts = my_run(circ)

plot_histogram(counts)
```
